# **Vector Genetic Diversity**



**What is genetic diversity**


Genetic diversity, the spectrum of genetic variation present within a species, is a fundamental component for population survival and adaptation in response to environmental changes. It provides a species with a diverse array of traits, enhancing its ability to thrive in different conditions.
As genetic diversity is crucial for a strong population, it maintains different genes that could lead to for exemple resistance to pests, diseases, or other stress conditions. It also enables individuals to adapt to various biotic and abiotic stresses. Among species, Anopheles mosquitoes stand out for their remarkable genetic diversity, surpassing many other organisms in this aspect.
However, genetic diversity can vary significantly among mosquito species and even within populations of the same species across different geographical locations and time periods. Exploring these variations is crucial as genetic diversity is influenced by demographic factors such as population size history and migration patterns. Therefore, studying and monitoring diversity can provide valuable insights into mosquito population demography, allowing for a deeper understanding of the demographic effects of vector control interventions.


**Different type of genetic diversity**

The genetic diversity arises from three main sources: mutation, recombination, and gene immigration. **Mutation** serves as the primary driver of genetic variation and evolution. Specifically, there are three types of DNA mutations: base substitutions (or point mutations), deletions, and insertions.

**Base substitutions or point mutations**

In the context of point mutations, there are three main categories:

1. Silent mutations: These occur when a nucleotide substitution does not alter the amino acid sequence in the polypeptide.

2. Missense mutations: These involve a change in the amino acid sequence. They can further be classified as conservative, where the change results in an amino acid from the same physical-chemical group, or non-conservative, where the change results in an amino acid from a different physicochemical group.

3. Non-sense mutations: These mutations result in the introduction of a premature stop codon due to a nucleotide modification.

**Frameshift mutations or insertions and deletions**

Nucleotide deletions and insertions are mutations that change the number of DNA base in genome. Deletions remove DNA segments, while insertions add DNA fragments, often leading to non-functional proteins. When deletions or insertions disrupt the gene's reading frame, impacting protein structure and function they are called frameshift mutations.

**Copy number variations**

Copy number variations (CNVs) in the form of large insertions and deletions were reported among the first genetic “mutations” ever. They are molecular phenomenon where sequences of the genome are repeated, and the number of repeats varies among individuals of the same species. CNVs play diverse biological roles, impacting physiological traits, morphological variations, metabolic states, susceptibility to insecticides, and host-microbiome interactions. They can significantly contribute to both common and rare genetic disorders. CNVs have the potential to influence mosquitoes population diversity and play a role in micro- and macro-evolutionary processes, which may have implications for the population dynamics.

## **How to quantify genetic diversity**

Many statistitics are used to evaluate how divergent individuals in a population are.

**S, Sn or Segragating site:** It refers to the number of polymorphic or segregating site. S depends on the sample size as well as the sequences size. The more the samples the more the segregating sites and the longer a sequence is the more segregating site could occur.

The **Watterson's estimator ( theta_w )** is the statisitics for that. It quantifies the number of segregating sites within a population, scaled by a constant. The higher this statistic, the more sites are found to be segregating (i.e., polymorphic).


**Nucleotide diversity (theta_pi)**: This statistics quantifies the average number of differences (mismatches) between any pair of DNA sequences in a population. The higher this statistic, the more differences you will expect to see between any pair of randomly sampled DNA sequences.

**Tajima's D (tajima_d)** This statistics quantifies the difference between Watterson’s estimator and nucleotide diversity, and scales the result by a constant. This statistic is useful because it focuses not on the magnitude but on the architecture of diversity in a cohort, compared to a neutrally evolving population of constant size. Negative values imply an excess of rare alleles, and positive values imply a deficit of rare alleles, relative to a population at equilibrium.
